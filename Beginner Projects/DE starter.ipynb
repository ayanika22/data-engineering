```python
import psycopg2 as psy 
import pandas as pd
```

## Create A Database

```python
def create_db():
    #create default database
    conn = psy.connect("host=localhost dbname=postgres user=postgres password=root")
    conn.set_session(autocommit=True)
    cur=conn.cursor()
    
    cur.execute("drop database if exists users")
    cur.execute("create database users")
    
    #close connection
    conn.close()
    
    conn = psy.connect("host=localhost dbname=users user=postgres password=root")
    conn.set_session(autocommit=True)
    cur=conn.cursor()
    
    return cur,conn
```

### Import and Clean Datasets


```python
url_path_0 = "https://drive.google.com/uc?id=1tAJd1Piu9ZSDis12WfJ8oU4CiB2F19zy"
ac_country = pd.read_csv(url_path_0)
ac_country.head()
```

```python
ac_country_clean = ac_country[['Code','Long Name','Region','Table Name','Short Name']]
ac_country_clean.head()
```

```python
url_path = "https://drive.google.com/uc?id=1E3M0j63FhC0bsQ4Wm59WiOOAU_lgtPvt"
ac_data = pd.read_csv(url_path)
ac_data.head()
```

```python
ac_data.columns
ac_data_clean = ac_data.loc[:,['Code', 'Topic', 'Indicator Name']]
```

```python
url_path_2 = "https://drive.google.com/uc?id=1VeykgbX_6ejqu-UuBaOcTD4TjXCQxgE3"
ac_series = pd.read_csv(url_path_2)
ac_series.head()
```

```python
ac_series_clean = ac_series.loc[:,['Country Name','Country Code','Series Code', '1995 [YR1995]','1996 [YR1996]', '1997 [YR1997]','1998 [YR1998]','1999 [YR1999]','2000 [YR2000]']]
```

```python
cur, conn = create_db()
```

### Create Tables

```python
ac_country_table_create = ("""CREATE TABLE IF NOT EXISTS ac_country(
country_code varchar primary key,
name varchar,
region varchar,
table_name varchar,
short_name varchar
)""")
```

```python
cur.execute(ac_country_table_create)
conn.commit()
```

```python
ac_data_table_create = ("""CREATE TABLE IF NOT EXISTS ac_data(
code varchar primary key,
topic varchar,
indicator_name varchar
)""")
```

```python
cur.execute(ac_data_table_create)
conn.commit()
```

```python
ac_series_table_create = ("""CREATE TABLE IF NOT EXISTS ac_series(
country_name varchar,
country_code varchar primary key,
series_code varchar,
Y1995 varchar,
Y1996 varchar,
Y1997 varchar,
Y1998 varchar,
Y1999 varchar,
Y2000 varchar
)""")
```

```python
cur.execute(ac_series_table_create)
conn.commit()
```

### Create Insert Queries

```python
ac_country_table_insert = """
    INSERT INTO ac_country (
        country_code,
        name,
        region,
        table_name,
        short_name
    )
    VALUES (%s, %s, %s, %s, %s) 
    ON CONFLICT (country_code) DO NOTHING
"""
```

```python
ac_data_table_insert = """
    INSERT INTO ac_data (code, topic, indicator_name)
    VALUES (%s, %s, %s)
    ON CONFLICT (code) DO NOTHING
"""
```

```python
ac_series_table_insert = """
    INSERT INTO ac_series (
    country_name, 
    country_code, 
    series_code, 
    Y1995, Y1996, Y1997, Y1998, Y1999, Y2000)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON CONFLICT (country_code) DO NOTHING
"""
```

### Insert data into tables

```python
for i, row in ac_country_clean.iterrows():
    cur.execute(ac_country_table_insert,list(row))
```

```python
for i, row in ac_data_clean.iterrows():
    cur.execute(ac_data_table_insert,list(row))
```

```python
for i, row in ac_series_clean.iterrows():
    cur.execute(ac_series_table_insert,list(row))
```
